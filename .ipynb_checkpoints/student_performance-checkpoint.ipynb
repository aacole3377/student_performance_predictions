```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import  DecisionTreeRegressor
from sklearn.metrics import precision_score,roc_auc_score,mean_squared_error,d2_absolute_error_score,accuracy_score
import warnings
```

```python
'''
Dataset Features:
StudentID: A unique identifier for each student.
Name: The name of the student.
Gender: The gender of the student (Male/Female).
AttendanceRate: The percentage of classes attended by the student.
StudyHoursPerWeek: The number of hours the student spends studying each week.
PreviousGrade: The grade the student achieved in the previous semester (out of 100).
ExtracurricularActivities: The number of extracurricular activities the student is involved in.
ParentalSupport: A qualitative assessment of the level of support provided by the student's parents (High/Medium/Low).
FinalGrade: The final grade of the student (out of 100), which serves as the target variable for prediction.
'''
```

```python
df = pd.read_csv("student_performance.csv")
df
```

```python
bins = [0, 60, 70, 80, 90, 100]
labels = ['F', 'D', 'C', 'B', 'A']
df['GradeCategory'] = pd.cut(df['FinalGrade'], bins=bins, labels=labels, right=False)
```

```python
plt.figure(figsize=(8, 6))
df['GradeCategory'].value_counts(sort=False).plot(kind='bar', color='skyblue')
plt.title('Histogram of Final Grades')
plt.xlabel('Grade')
plt.ylabel('Number of Students')
plt.xticks(rotation=0)  # Rotate labels on x-axis
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['AttendanceRate'], df['FinalGrade'], color='blue')
plt.xlabel('Attendance Rate')
plt.ylabel('Final Grade')
plt.xticks(rotation=0)
plt.show()
```

```python
'''AttendanceRate                int64
StudyHoursPerWeek             int64
PreviousGrade                 int64
ExtracurricularActivities     int64
ParentalSupport              object
FinalGrade                    int64
'''

grouped_gender = df.groupby('Gender')[['AttendanceRate', 'StudyHoursPerWeek', 'PreviousGrade', 'ExtracurricularActivities', 'FinalGrade']].mean()
grouped_gender
```

```python
dummies = pd.get_dummies(df, columns=['Gender', 'ParentalSupport'])
dummies = dummies.drop(columns=['StudentID' , 'Name', 'GradeCategory'])
dummies = dummies.astype(int)
dummies.info()
dummies
```

```python
X = dummies.drop('FinalGrade', axis=1)
Y = dummies['FinalGrade']
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
randomForestRegressor = RandomForestRegressor()
model = randomForestRegressor.fit(x_train, y_train)
model.score(x_test, y_test)
```

```python
prediction = randomForestRegressor.predict(x_test)
mse = mean_squared_error(y_test, prediction)
rmse = mse**.5
print(mse)
print(rmse)
```
